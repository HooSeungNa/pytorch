```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_files='./vid_images/'
if not os.path.exists(image_files):
    os.makedirs(image_files)
    print("folders created!")
else:
    print("folders not created!")
```

```python
vidcap=cv2.VideoCapture('./video1.mp4')
cnt=0
```

```python
length = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
width = int(vidcap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = vidcap.get(cv2.CAP_PROP_FPS)

print('length : ',length)
print('width : ',width)
print('height : ',height)
print('fps : ',fps)

while(vidcap.isOpened()):
    ret,image=vidcap.read()
    d_img=np.rot90(image)
#     cv2.imwrite(image_files+'before_frame%d.jpg'%(cnt+1),image)
    cv2.imwrite(image_files+'after_frame%d.jpg'%(cnt+1),d_img)
    print('saved frame%d.jpg'%(cnt+1))
    cnt+=1
    if(int(vidcap.get(1))==length):
        break
vidcap.release()
```

```python
from icrawler.builtin import GoogleImageCrawler
import os
download_dir="D://image_file/cat"
if not os.path.exists(download_dir):
    os.makedirs(download_dir)
    print("make directory")
```

```python
google_crawler=GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=1,
    downloader_threads=4,
    storage={'root_dir':download_dir})
filters=dict(size='large', # 
            color='black',#  
            license='commercial,modify',
            date=((2017,1,1),(2017,12,31)))#   

google_crawler.crawl(keyword='cat',
                     filters={'date':((2016,1,1),(2016,6,30))},
                     max_num=1000,
                     file_idx_offset=0)
google_crawler.crawl(keyword='cat',
                     filters={'date':((2016,7,1),(2016,12,31))},
                     max_num=1000,
                     file_idx_offset='auto')
google_crawler.crawl(keyword='cat',
                     filters={'date':((2017,1,1),(2017,6,30))},
                     max_num=1000,
                     file_idx_offset='auto')
google_crawler.crawl(keyword='cat',
                     filters={'date':((2017,7,1),(2017,12,31))},
                     max_num=1000,
                     file_idx_offset='auto')

```

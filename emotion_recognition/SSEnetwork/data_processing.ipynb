```python
import cv2
import numpy as np
import glob
import os
from xml.etree.ElementTree import parse, dump
import csv
import shutil
```

```python
dirname="D:/emotion_dataset/mmidataset/VideoWithImageLabels/Sessions/"
```

```python
filenames = os.listdir(dirname)
filelist=[]
full_filename=[]
for filename in filenames:
    a= os.path.join(dirname, filename)
    full_filename.append(a)
    if(len(os.listdir(a))==4):
        fname, ext = os.path.splitext(os.listdir(a)[2])
    else:
        fname, ext = os.path.splitext(os.listdir(a)[1])
```

```python
#file list avi list generate
filelist=[]
avilist=[]
for filename in filenames:
    file_tmp = os.path.join(dirname, filename)
    avi_tmp = os.path.join(dirname, filename)
    
    if(len(os.listdir(file_tmp))==4):
        file_tmp=file_tmp+'/'+os.listdir(file_tmp)[2]
        avi_tmp=avi_tmp+'/'+os.listdir(avi_tmp)[1]
    else:
        file_tmp=file_tmp+'/'+os.listdir(file_tmp)[1]
        avi_tmp=avi_tmp+'/'+os.listdir(avi_tmp)[0]
    filelist.append(file_tmp)
    avilist.append(avi_tmp)
print("how many xml files : ",len(filelist))
print("how many avi files : ",len(avilist))
```

```python
#6개의 감정만 남겨놓고 나머지는 삭제 한다.
e0=0
e1=0
e2=0
e3=0
e4=0
e5=0
e6=0
e7=0
e8=0
e9=0
e10=0
e11=0
e12=0
e13=0
e14=0
e15=0

f = open('D:/emotion_dataset/mmidataset/VideoWithImageLabels/labels.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(f)
for i in range(len(filelist)):
    tree=parse(filelist[i])
    for item in tree.findall('Metatag'):
        a=item.attrib
        if(a['Name']=='Emotion'):
            e_value=int(a['Value'])
            wr.writerow([avilist[i],e_value])
            if(e_value==0):
                e0+=1
            elif(e_value==1):
                e1+=1
            elif(e_value==2):
                e2+=1
            elif(e_value==3):
                e3+=1
            elif(e_value==4):
                e4+=1
            elif(e_value==5):
                e5+=1
            elif(e_value==6):
                e6+=1
            elif(e_value==7):
                shutil.rmtree(full_filename[i])
            elif(e_value==8):
                shutil.rmtree(full_filename[i])
            elif(e_value==9):
                shutil.rmtree(full_filename[i])
            elif(e_value==11):
                shutil.rmtree(full_filename[i])
            elif(e_value==12):
                shutil.rmtree(full_filename[i])
            elif(e_value==13):
                shutil.rmtree(full_filename[i])
            elif(e_value==14):
                shutil.rmtree(full_filename[i])


f.close()
    
```

```python
#video to image and labeling 
image_dir='D:/emotion_dataset/mmidataset/VideoWithImageLabels/image_dataset'
#generate image file 
if not(os.path.exists(image_dir)):
    os.mkdir(image_dir)
else:
    print("dir already exist")
```

```python

write_label = open('D:/emotion_dataset/mmidataset/VideoWithImageLabels/image_dataset/labels.csv'
                   ,'w', encoding='utf-8', newline='')
wr = csv.writer(write_label)

#csv 파일을 통해서 
read_label= open('D:/emotion_dataset/mmidataset/VideoWithImageLabels/labels.csv', 'r', encoding='utf-8')
rdr = csv.reader(read_label)
i=0
for line in rdr:
    print(line[0])
    vidcap=cv2.VideoCapture(line[0])
    cnt=0
    length = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(vidcap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = vidcap.get(cv2.CAP_PROP_FPS)
    print('length : ',length)
    print('width : ',width)
    print('height : ',height)
    print('fps : ',fps)

    while(vidcap.isOpened()):
        ret,image=vidcap.read()
        d_img=np.rot90(image)
        cv2.imwrite(image_dir+'/'+'image%d_frame%d.jpg'%((i),(cnt+1)),d_img)
        wr.writerow(['image%d_frame%d.jpg'%((i),(cnt+1)),int(line[1])])
        print('saved frame%d.jpg'%(cnt+1))
        cnt+=1
        if(int(vidcap.get(1))==length):
            break
    vidcap.release()
    i+=1
read_label.close()    
write_label.close()


```
